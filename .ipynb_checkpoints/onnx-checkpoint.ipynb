```python
def letterbox(im, new_shape=(1280, 1280), color=(114, 114, 114), auto=True, scaleup=True, stride=32):
    # Resize and pad image while meeting stride-multiple constraints
    shape = im.shape[:2]  # current shape [height, width]
    if isinstance(new_shape, int):
        new_shape = (new_shape, new_shape)

    # Scale ratio (new / old)
    r = min(new_shape[0] / shape[0], new_shape[1] / shape[1])
    if not scaleup:  # only scale down, do not scale up (for better val mAP)
        r = min(r, 1.0)

    # Compute padding
    new_unpad = int(round(shape[1] * r)), int(round(shape[0] * r))
    dw, dh = new_shape[1] - new_unpad[0], new_shape[0] - new_unpad[1]  # wh padding

    if auto:  # minimum rectangle
        dw, dh = np.mod(dw, stride), np.mod(dh, stride)  # wh padding

    dw /= 2  # divide padding into 2 sides
    dh /= 2

    if shape[::-1] != new_unpad:  # resize
        im = cv2.resize(im, new_unpad, interpolation=cv2.INTER_LINEAR)
    top, bottom = int(round(dh - 0.1)), int(round(dh + 0.1))
    left, right = int(round(dw - 0.1)), int(round(dw + 0.1))
    im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)  # add border
    return im, r, (dw, dh)
```

```python
def generate_bounding_box(image_path, bounding_boxes):
    label_colors = {
        2: (0, 170, 0),           # Good
        5: (0, 0, 170),           # Skinout
        4: (85, 0, 85),           # Shrivelled
        3: (170, 0, 0),           # Insectbite
        1: (0, 119, 170),         # Discolored
        0: (250, 0, 255),         # Broken
        6: (0, 170, 170)          # Sprouted
    }
    circle_radius = 5
    img = cv2.imread(image_path)
    image_height, image_width, _ = img.shape
    new_shape = (max(image_height, image_width), max(image_height, image_width))
    image, ratio, dwdh = letterbox(img, new_shape=new_shape, auto=False)
    i = 0

    # Check the format of bounding_boxes
    print(f"Bounding boxes type: {type(bounding_boxes)}, length: {len(bounding_boxes)}")
    print(f"Example bounding box: {bounding_boxes[0]}")

    # Extract bounding box coordinates and class from the array
    for bounding_box in bounding_boxes:
        # Print each bounding box for debugging
        print(f"Processing bounding box: {bounding_box}")

        # Ensure bounding_box is iterable and contains 5 elements
        if len(bounding_box) != 5:
            print(f"Unexpected bounding box format: {bounding_box}")
            continue  # Skip this bounding box

        x_min, y_min, x_max, y_max, class_id = bounding_box
        box = np.array([x_min, y_min, x_max, y_max])
        box = box.round().astype(np.int32).tolist()

        # Define colors based on class_id
        circle_color = label_colors.get(class_id, (255, 255, 255))
        text_class = str(i)

        # Draw the bounding box on the image
        cv2.rectangle(image, box[:2], box[2:], circle_color, 2)
        cv2.putText(image, str(i), (box[0], box[1] - 2), cv2.FONT_HERSHEY_SIMPLEX, 0.30, [225, 255, 255], thickness=1)
        i += 1

    output_path = image_path + "_output.jpg"
    cv2.imwrite(output_path, image)
    print(f"Saved output image to {output_path}")
```

```python
import os
import cv2
import numpy as np
import tritonclient.grpc as grpcclient
import time

# Initialize the Triton client
triton_client = grpcclient.InferenceServerClient(url="13.234.216.76:8001")

INPUT_NAMES = ["raw"]
OUTPUT_NAMES = ["results", "boxclass"]

outputs = [
    grpcclient.InferRequestedOutput(OUTPUT_NAMES[0]),
    grpcclient.InferRequestedOutput(OUTPUT_NAMES[1])
]

images = os.listdir("./FM2/")

# Record the start time
start_time = time.time()

for img_path in images:
    img_path = os.path.join("./FM2", img_path)
    input_image = cv2.imread(img_path)
    orig_image = input_image.copy()

    # Create a fresh InferInput object for each image
    infer_input = grpcclient.InferInput(INPUT_NAMES[0], [1, input_image.shape[0], input_image.shape[1], 3], "UINT8")
    input_image_buffer = np.expand_dims(orig_image, axis=0)
    infer_input.set_data_from_numpy(input_image_buffer)
    
    # Prepare the inputs list for this iteration
    inputs = [infer_input]
    
    # Perform inference
    triton_results = triton_client.infer('ensemble', inputs=inputs, outputs=outputs)
    
    # Process the results
    output0_data = triton_results.as_numpy("boxclass")
    data = output0_data.squeeze()
    print("output0_data ", data.shape)
    
    output1_data = triton_results.as_numpy("results")
    prediction = output1_data.squeeze()
    print("output1_data ", prediction)
    
    generate_bounding_box(img_path, data)

# Calculate and print the total time taken for inference
total_time = time.time() - start_time
print(f"Total time taken for inference: {total_time:.2f} seconds")
```
